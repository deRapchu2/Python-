```python
import uuid 
class Customer:  # Syntax for defining a class
    
    # Class Attribute:
    # A class attribute is shared across all instances of a class.
    # It can be used to identify the class or provide a common property.
    type = 'customer'  
    
    # Constructor (__init__):
    # Used to initialize instance attributes (unique to each object).
    # Python allows only one __init__ per class (no function overloading).
    def __init__(self, nam, mbn, adr):  
        # Instance Attributes:
        # Attributes defined with self belong to the object (instance), not the class.
        # By default, they are public (can be accessed and modified outside the class).
        # If prefixed with '__' (double underscore), they become private.
        self.name = nam
        self.mobile = mbn
        self.address = adr
        self.id = uuid.uuid4()
    
    # String Representation (__str__):
    # Defines how the object is represented when printed.
    def __str__(self):
        return f"Mr. {self.name} lives at {self.address} and you can contact them on {self.mobile}\n"
    
    # Encapsulation:
    # By combining attributes and methods inside the same class,
    # we achieve encapsulation (one of the OOP principles).


# Example usage:

# Creating an instance using the parameterized constructor:
test3 = Customer('Zaid', '2423456', 'Baramulla')

# Accessing instance attributes directly:
print(test3.name)

# Printing the object automatically calls __str__:
print(test3)

# Accessing class attribute:
print(test3.type)

# Accessing instance attribute again:
print(test3.name)

# Notes on Class Attributes:
# If we modify a class attribute, it is reflected across all instances.
# However, if an instance attribute with the same name exists,
# it overrides the class attribute for that instance.

# Example:
# test2 = Customer('Ali', '9999999', 'Srinagar')
# test2.type = 'premium_customer'   # This will override only for test2
# print(test3.type)  # Still 'customer'
# print(test2.type)  # 'premium_customer'
```
